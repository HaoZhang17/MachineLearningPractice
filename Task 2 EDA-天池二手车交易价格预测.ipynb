## 导览

探索性数据分析(Exploratory Data Analysis, EDA)是分析中非常基础且重要的一步，该步骤有助于分析者对数据的大小、组成、分布等属性产生初步的认识，同时了解变量与变量、变量与预测值之间的相关性，对数据是否可用于后续分析做出判断，并对后续建模步骤的参数选择有大致的见解。

## 经验
1. 注意缺失数据的具体表现形式，类别型特征（非数值型特征）的缺失数据的表现形式在不同数据集间可能存在差异。在结合isnull等函数判断缺失值时，需再结合value_counts等统计函数进一步确认类别型数据是否含有缺失值；
2. 对含有异常值的特征列进行分析时，应注意异常值是否应该剔除，或用正常值填充；
3. 在数据分析前，应了解数据的产生背景、方法和针对数据产生的已有的结论，思考被纳入分析的特征的合理性、可能存在异常值之处及产生原因。数据分析过程中，应对数据有较高的敏感性，随时注意产生的结果，不忽视任何异常现象。数据初步分析完成后，对数据进行评估。

### 3.1.1 总体分布情况

```python
import scipy.stats as st
y = Train_data['price']
# 不符合正态分布
plt.figure(2); plt.title('Normal')
sns.distplot(y, kde=False, fit=st.norm)
# 对数转换
plt.figure(3); plt.title('Log Normal')
sns.distplot(y, kde=False, fit=st.lognorm)
# 最佳拟合
plt.figure(1); plt.title('Johnson SU')
sns.distplot(y, kde=False, fit=st.johnsonsu)
```

### 4.2.2 类别特征箱线图可视化

```python
# name和regionCode的类别太稀疏，这里将不稀疏的几类画一下
categorical_features = ['model',
                        'brand',
                        'bodyType',
                        'fuelType',
                        'gearbox',
                        'notRepairedDamage']
for c in categorical_features:
    Train_data[c] = Train_data[c].astype('category')
    if Train_data[c].isnull().any():
        # 增加MISSING类别
        Train_data[c] = Train_data[c].cat.add_categories(['MISSING'])
        # 将NaN用MISSING填充
        Train_data[c] = Train_data[c].fillna('MISSING')

def boxplot(x, y, **kwargs):
    sns.boxplot(x=x, y=y)
    x=plt.xticks(rotation=90)

f = pd.melt(Train_data, id_vars=['price'], value_vars=categorical_features)
g = sns.FacetGrid(f, col="variable",  col_wrap=2, sharex=False, sharey=False, size=5)
g = g.map(boxplot, "value", "price")
```
